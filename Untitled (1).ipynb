```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("Students Social Media Addiction.csv")

df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()        # Remove extra spaces
df.columns = df.columns.str.replace(" ", "_")  # Replace spaces with _
df.columns = df.columns.str.lower()       # Convert to lowercase

df.columns
```

```python
df.head()
```

```python
sns.histplot(df['avg_daily_usage_hours'], kde=True)
plt.show()
```

```python
sns.countplot(x='most_used_platform', data=df)
plt.xticks(rotation=45)
plt.title("Most Used Social Media Platform")
plt.show()
```

```python
sns.scatterplot(x='avg_daily_usage_hours', y='affects_academic_performance', data=df)
plt.title("Usage Hours vs Academic Performance")
plt.show()
```

```python
sns.countplot(x='academic_level', data=df)
plt.title("Academic Level Distribution")
plt.show()
```

```python
df['Usage_Category'] = df['avg_daily_usage_hours'].apply(
    lambda x: 'Low' if x < 2 else 'Medium' if x < 5 else 'High'
)
```

```python
plt.figure(figsize=(10,6))

numeric_df = df.select_dtypes(include=['int64', 'float64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")

plt.show()
```
